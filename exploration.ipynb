```python
audioldm = None
current_model_name = None

model_name = 'audioldm_16k_crossattn_t5' # doesnt work very well?
# model_name = 'audioldm2-full' # Needs more memory than what I have
# model_name = 'audioldm_48k'

audioldm=build_model(model_name=model_name)
```

```python
def text2audio(text, duration, guidance_scale, random_seed, n_candidates, model_name):
    global audioldm, current_model_name
    torch.set_float32_matmul_precision("high")

    if("48k" in model_name):
        latent_t_per_second=12.8
        sample_rate=48000
    else:
        latent_t_per_second=25.6
        sample_rate=16000

    
    # print(text, length, guidance_scale)
    waveform = text_to_audio(
        latent_diffusion=audioldm,
        text=text,
        seed=random_seed,
        duration=duration,
        guidance_scale=guidance_scale,
        n_candidate_gen_per_text=int(n_candidates),
        latent_t_per_second=latent_t_per_second,
    )  # [bs, 1, samples]
    # waveform = [
    #     gr.make_waveform((16000, wave[0]), bg_image="bg.png") for wave in waveform
    # ]
    # waveform = [(16000, np.random.randn(16000)), (16000, np.random.randn(16000))]
    if len(waveform) == 1:
        waveform = waveform[0]
    return waveform
```

```python
text = "A drumstick is continuously hitting a hard metal surface in a large room."
duration = 10.0 #Duration is minimum 10 secs. The generated sounds are weird for <10secs
guidance_scale = 3
random_seed = 45
n_candidates = 3

wav = text2audio(text, duration, guidance_scale, random_seed, n_candidates, model_name)
print(len(wav))
IPython.display.display(IPython.display.Audio(wav[0], rate=16000))
```
